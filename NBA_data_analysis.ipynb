```python
df = df.dropna()
df.tail()
```

```python
def height_to_meters(height_str):
    #print(type (height_str))
    height_str=height_str.split('-')
    
    feet, inches = map(int, height_str)
    #print(height_str)
    height_in_inches = feet * 12 + inches
    height_in_meters = height_in_inches * 0.0254
    print(height_in_meters)
    return (height_in_meters)
```

```python
df['height_in_meters'] = df['Height'].apply(height_to_meters)

```

```python
def weight_to_kilos(weight_pounds):
    return weight_pounds * 0.453592


```

```python
df['Weight_in_kilos'] = df['Weight'].apply(weight_to_kilos)
print(df)
```

```python
print(df.describe())
```

```python
#Height vs Weight Scatter Plot
plt.figure(figsize=(15, 10))
sns.scatterplot(data=df, x='height_in_meters', y='Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
# Salary Distribution
plt.figure(figsize=(15, 10))
sns.boxplot(data=df, x='Position', y='Salary')
plt.title('Salary Distribution by Position')
plt.show()
```

```python
#  Age vs Salary Scatter Plot
plt.figure(figsize=(15, 10))
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Age vs Salary')
plt.show()
```

```python
#  College Representation (how many players from each college)
college_count = df['College'].value_counts()
print(f"College Representation:\n{college_count}")
```

```python
#Handling missing salary values
missing_salary_count = df['Salary'].isna().sum()
print(f"Missing Salary Count: {missing_salary_count}")
```

```python
#Correlation between Age and Salary
correlation = df[['Age', 'Salary']].corr()
print(f"Correlation between Age and Salary:\n{correlation}")
```

```python
#Histogram for Age Distribution
plt.figure(figsize=(15, 10))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution of Players')
plt.show()
```
